**Clean the Column Names**

*   Removing extra whitespace.
*   Converting to lowercase.
*   Replacing spaces with underscores.
*   Removing special characters.


```python
# Clean the column names by removing spaces, converting to lowercase, and replacing special characters
df.columns = (
    df.columns
    .str.strip()                              # Remove leading and trailing whitespace
    .str.lower()                              # Convert all characters to lowercase
    .str.replace(' ', '_')                    # Replace spaces with underscores
    .str.replace(r'[^\w_]', '', regex=True)   # Remove special characters
)
```

**Handle Missing Values**

```python
df.isnull().sum()
```

For the electric_range column, the mean is calculated and used to fill missing values.


```python
mean_value = df['electric_range'].mean()

df['electric_range'] = df['electric_range'].fillna(mean_value)
```

For the base_msrp (Manufacturer Suggested Retail Price) column, the median is used to fill missing values. This method is suitable for price-related data, as the median reduces the influence of outliers.

```python
median_value = df['electric_range'].median()

df['base_msrp'] = df['base_msrp'].fillna(mean_value)
```

The data is grouped by the 'model' column. This means that the dataset is partitioned into smaller groups, where each group consists of rows sharing the same vehicle model.

```python
df['legislative_district'] = df.groupby('model')['legislative_district'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Convert postal_code to string and remove the decimal
df['postal_code'] = df['postal_code'].astype(int).astype(str)
```

```python
# Convert model_year to integer type (if not already)
df['model_year'] = df['model_year'].astype(int)
```

**Convert Columns to Numeric Types**

```python
# Convert electric_range and base_msrp to numeric types
df['electric_range'] = pd.to_numeric(df['electric_range'], errors='coerce')
df['base_msrp'] = pd.to_numeric(df['base_msrp'], errors='coerce')
```

```python
# Define a function to extract latitude and longitude from the 'POINT' string
def extract_lat_long(point_str):
    point_str = point_str.replace('POINT (', '').replace(')', '')
    longitude, latitude = map(float, point_str.split())
    return latitude, longitude

# Apply the function and create new columns for latitude and longitude
df[['latitude', 'longitude']] = df['vehicle_location'].apply(lambda x: pd.Series(extract_lat_long(x)))
```
